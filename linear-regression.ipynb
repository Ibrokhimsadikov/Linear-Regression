```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor


from sklearn.metrics import mean_squared_error
```

```python
#read your csv files here
day = pd.read_csv('day.csv')
hour =pd.read_csv('hour.csv') 
```

## Hour Dataset

### Q1(a) number of casual riders

```python
#separating predictors and target columns: target here is 'casual' column
#drop columns ['casual', 'registered', 'cnt','dteday','instant'] from X
X = hour.drop(columns=['casual', 'registered', 'cnt', 'dteday', 'instant'])
y = hour[['casual']]
```

```python
#print the shape of X and y
print(X.shape)
print(y.shape)
```

```python
#do a train test split, use test_size =0.3

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)



print("Linear Regression")
#initialize your linear regression model
lr = LinearRegression()
#fit training set to your LR model
lr = lr.fit(X_train, y_train)
#predict on the test set using your model
y_pred = lr.predict(X_test)
#evaluate: use MSE
print("MSE for LR: %.2f" %mean_squared_error(y_test,y_pred))




print("KNN")
#initialize your KNearestNeighbors model
knn = KNeighborsRegressor()
#fit training set to your KNN model
knn = knn.fit(X_train, y_train)
#predict on the test set using your model
y_pred = knn.predict(X_test)
#evaluate: use MSE
print("MSE for KNN: %.2f" %mean_squared_error(y_test,y_pred))






```
